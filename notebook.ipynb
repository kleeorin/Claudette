```python
# The Mandelbrot set in pure stdlib Python — z -> z² + c, rendered in ASCII
def mandelbrot(width=76, height=30, max_iter=40):
    chars = " .:-=+*#%@"
    for row in range(height):
        y = 1.15 - 2.3 * row / (height - 1)
        line = ""
        for col in range(width):
            x = -2.2 + 3.0 * col / (width - 1)
            z, c, i = 0j, complex(x, y), 0
            while abs(z) <= 2 and i < max_iter:
                z = z * z + c
                i += 1
            line += chars[9 * i // max_iter]
        print(line)

mandelbrot()
```
